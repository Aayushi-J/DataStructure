1) Enter 10000 elements in an array and verify the all types of sorting methods


```python
import random
import time
import matplotlib.pyplot as plt
def bubble_sort(list1):
    # Outer loop for traverse the entire list
    for i in range(0,len(list1)-1):
        for j in range(len(list1)-1):
            if(list1[j]>list1[j+1]):
                temp = list1[j]
                list1[j] = list1[j+1]
                list1[j+1] = temp
    return list1
def insertionSort(array):
    for step in range(1, len(array)):
        key = array[step]
        j = step - 1
        while j >= 0 and key < array[j]:
            array[j + 1] = array[j]
            j = j - 1
        array[j + 1] = key
    return array
def selectionSort(array):
    for ind in range (0,len(array)-1):
        min_index = ind
        for j in range(ind + 1, len(array)):
            if array[j] < array[min_index]:
                min_index = j
        (array[ind], array[min_index]) = (array[min_index], array[ind])
    return array
search=[]
search1=[]
search2=[]
for i in range(0,1000):
    s=random.randint(1,10000)
    search.append(s)
    search1.append(s)
    search2.append(s)
timeb=[]
for i in search:
    start=time.time_ns()
    w=bubble_sort(search)
    end=(time.time_ns()-start)
    timeb.append(end)
timei=[]
for i in search:
    start1=time.time_ns()
    w=insertionSort(search1)
    end1=(time.time_ns()-start1)
    timei.append(end1)
times=[]
for i in search:
    start2=time.time_ns()
    w=selectionSort(search2)
    end2=(time.time_ns()-start2)
    times.append(end2)
print(timeb)
print(timei)
print(times)
plt.plot(search,timeb,color = "red")
plt.show()
print("\n")
plt.plot(search1,timei,color = "blue")
plt.show()
print("\n")
plt.plot(search2,times,color = "green")
plt.show()
```

2)Given m sorted array/lists, each containing n elements print them efficiently from the sorted order



```python
 import heapq
def sorted_lists(arrays):
    # Min heap for the selection of the most smallest element
    min_heap = []

    # putting elements into the min heap according to the index value
    for i in range(len(arrays)):
        if len(arrays[i]) > 0:
            heapq.heappush(min_heap, (arrays[i][0], i, 0))

    #all the smallest elements from each of the array enter into the heap
    while min_heap:
        val, array_index, element_index = heapq.heappop(min_heap)
        print(val)  # Print the smallest element

        # Move to the next element in the same array
        element_index += 1

        # Check if there are more elements in the current array
        if element_index < len(arrays[array_index]):
            heapq.heappush(min_heap, (arrays[array_index][element_index], array_index, element_index))

# Example usage:
array1 = [10, 20, 30, 40]
array2 = [15, 25, 35]
array3 = [27, 29, 37, 48, 93]
array4=[32,33]

arrays = [array1, array2, array3, array4]

sorted_lists(arrays)
```

3) Given an array of size N, find the K largest elemengts in the array where k<<<N

```python
import heapq
def largest(arr, k):
    if k >= len(arr):
        return arr
    heap = arr[:k]  # Create a min-heap of the first K elements
    heapq.heapify(heap)

    for num in arr[k:]:
        if num > heap[0]:  # Compare with the smallest element in the heap
            heapq.heappop(heap)
            heapq.heappush(heap, num)
    return heap
array=[]
for i in range(0,1000):
  s=random.randint(1,1000)
  array.append(s)
print(largest(array,4))
```

4) Given  a set of activities along with their starting and ending time , find the maximum number of activities performed by a single person assuming that a person can only work on a single activity at a time


```python
def max(array1):
    array1.sort(key=lambda x: x[1])
    count = 1
    prev = array1[0][1]
    for i in range(1, len(array1)):
        if array1[i][0] >= prev:
            count += 1
            prev = array1[i][1]
    return count
a = [(1, 4), (3, 5), (0, 6), (5, 7), (3, 8),(5, 9),(6, 10),(8, 11),(8, 12),(12, 13),(12,14)]
print(max(a))
```
